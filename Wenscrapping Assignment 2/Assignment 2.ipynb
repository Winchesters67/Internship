```python
#4
driver=webdriver.Chrome(r"G:\chromedriver.exe")
driver.get("https://www.flipkart.com/")

location=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input")
location.send_keys('sunglasses')

search=driver.find_element(By.CLASS_NAME,"_34RNph")
search.click()

brand_name=[]
Product_Description=[]
Price=[]
discount=[]
start=0
end=3

for page in range(start, end):
    brand=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand[0:100]:
        brand_name.append(i.text)

    PROD_DES=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in PROD_DES[0:100]:
        Product_Description.append(i.text)

    price=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in price[0:100]:
        Price.append(i.text)

    Discount=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in Discount[0:100]:
        discount.append(i.text)

    nextb=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    nextb.click()
    time.sleep(4)

brand_name=brand_name[0:100]
Product_Description=Product_Description[0:100]
Price=Price[0:100]
discount=discount[0:100]

df=pd.DataFrame({'Brand':brand_name,'Product Description':Product_Description,'Price':Price,'Discount':discount})
df
```
